```python
import pickle
import os

def saveDictionary(tmp_dict):
    try:
        with open('words.txt', 'wb') as dictionary:
            pickle.dump(tmp_dict, dictionary)
        print("SUCCESS: writing to dictionary words.txt")
    except Exception as e:
        print(f"ERROR: {e}")

def getDictionary():
    try:
        with open('words.txt', 'rb') as dictionary:
            tmp_dict = pickle.load(dictionary)
        print("SUCCESS: reading from dictionary words.txt")
        return tmp_dict
    except Exception as e:
        print(f"ERROR: {e}")

def addNewWord():
    print("__________")
    print("NEW WORD ADDER:")
    print("__________")
    print("Enter a Word to add")
    word = input(">>")
    print(f"Enter a meaning of {word} to add")
    meaning = input(">>")
    dictionary = getDictionary()
    dictionary[word] = meaning
    saveDictionary(dictionary)
    main_menu()

def findMeaning():
    print("__________")
    print("WORD FINDER:")
    print("__________")
    print("Enter a Word to find")
    word = input(">>").lower() 
    dictionary = getDictionary()
    
    found_entries = []
    
    for key in dictionary:
        if key.lower() == word:
            found_entries.append((key, dictionary[key]))

    if found_entries:
        # Print all found entries
        for entry in found_entries:
            print(f"Found Entry: [{entry[0]} : {entry[1]}]")
    else:
        print("ERROR: The word doesn't exist!")
    
    main_menu()


def updateWord():
    print("__________")
    print("WORD UPDATER:")
    print("__________")
    print("Enter an old Word to edit")
    word = input(">>")
    dictionary = getDictionary()
    if word in dictionary:
        print(f"Enter a New Meaning to edit [{word} : {dictionary[word]}] to:")
        new_meaning = input(">>")
        dictionary[word] = new_meaning
        saveDictionary(dictionary)
        print(f"SUCCESS: Updated to [{word} : {dictionary[word]}]")
    else:
        print("ERROR: The word doesn't exist! Impossible to update!")
        updateWord()
    main_menu()

def main_menu():
    if not os.path.exists(os.path.join("words.txt")):
        try:
            with open('words.txt', 'wb') as dictionary:
                pickle.dump({}, dictionary)
            print("SUCCESS: writing to dictionary words.txt")
        except Exception as e:
            print(f"ERROR: {e}")
    print("__________")
    print("ENGLISH DICTIONARY APP")
    print("__________")
    print("1. Add a new word")
    print("2. Find the meaning")
    print("3. Update a word")
    print("4. Exit")
    print("__________")
    n = int(input(">>"))
    if n == 1:
        addNewWord()
    elif n == 2:
        findMeaning()
    elif n == 3:
        updateWord()
    elif n == 4:
        exit()
    else:
        print("ERROR: Please Enter a digit according to Menu.")
        main_menu()

main_menu()
```
